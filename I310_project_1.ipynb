```python
import pandas as pd
import matplotlib.pyplot as plt

data1 = pd.read_csv ("greenhouse-gas-emissions2022.csv")

df = pd.DataFrame(data1, columns=['Anzsic', 'Anzsic_description', 'Period', 'Data_value', 'Variable', 'Units', 'Gas', 'Status'])

data_values = df['Data_value']

plt.hist(data_values, bins=20, edgecolor='k')
plt.xlabel('Total Emissions')
plt.ylabel('Number of Times Greenhouse Gasses Were Emitted')
plt.title('Greenhouse Gasses Emitted by All Companies')

plt.show()
```

This histogram provides the average of how much greenhouse gasses were put into the atmosphere by the different companies
Provides a visual of which type of companies emit more or less Greenhouse gasses.

```python
import pandas as pd
import matplotlib.pyplot as plt

data1 = pd.read_csv ("greenhouse-gas-emissions2022.csv")

df = pd.DataFrame(data1, columns=['Anzsic', 'Anzsic_description', 'Period', 'Data_value', 'Variable', 'Units', 'Gas', 'Status'])

periods = df['Period']
data_values = df['Data_value']

plt.scatter(periods, data_values, marker='o', alpha=0.5)

plt.xlabel('Each Quarter')
plt.ylabel('Total Emissions')
plt.title('Total Emissions by Year')

plt.xticks(rotation=45)

plt.show()
```

The scatter plot above provides a visual of which type of company emitted the most greenhouse gasses.
This visual provides a clear indication that the mining companies emit the least amount of Greenhouse gasses while the Agriculture, Forestry, Fishing and Primary companies emit the most Greenhouse gasses while creating their products.
